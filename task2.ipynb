```python
# Install required libraries if not already installed
!pip install seaborn matplotlib pandas

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset (adjust the filename if different)
df = pd.read_csv("dataset.csv")

# View top rows
df.head()
# Basic stats
df.describe()
# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df['Sleep_Hours'], bins=10, kde=True)
plt.title("Distribution of Sleep Hours")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()

# Boxplot
plt.figure(figsize=(6,2))
sns.boxplot(x=df['Sleep_Hours'], color="skyblue")
plt.title("Boxplot of Sleep Hours")
plt.show()
sns.pairplot(df[['Sleep_Hours', 'Hours_Studied', 'Depression_Score', 'Exam_Score']], diag_kind='kde')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
# Correlation Matrix
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sleep_Hours','Hours_Studied', 'Depression_Score', 'Exam_Score' ]].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(x='Stress_Level', y='Exam_Score', data=df, palette='Set2')
plt.title("Stress Level vs. Exam Score")
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Depression_Score', data=df, palette='pastel')
plt.title("Gender vs. Depression Score")
plt.show()
```
